```python
#Using SimpleImputer from Scikit-Learn Library
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer 
datasets = pd.read_csv('/content/Exercise-CarData.csv') 
print("\nData :\n",datasets)
```

```python
datasets['KM']=datasets['KM'].replace("??",np.nan)
datasets['HP']=datasets['HP'].replace("????",np.nan)
datasets['Doors']=datasets['Doors'].replace("three",'3')
datasets['Doors']=datasets['Doors'].replace("four",'4')
datasets['Doors']=datasets['Doors'].replace("five",'5')
datasets=datasets.astype({"Doors":int,"KM":float,"HP":float})
datasets.dropna(axis=0,how="all",inplace=True)
print("\nData :\n",datasets)
print(datasets.info())
```

```python
datasets['FuelType']=datasets['FuelType'].fillna(datasets['FuelType'].mode()[0])
new_x1=datasets.loc[:, datasets.columns != 'FuelType'].values
imputer = SimpleImputer(missing_values = np.nan,strategy = "mean")
imputer = imputer.fit(new_x1[:,1:]) 
new_x1[:,1:] = imputer.transform(new_x1[:,1:]) 
print(new_x1[:,1:]) 
```
